### 開始做集群分析

```python
from sklearn.preprocessing import scale
from sklearn import cluster
```

```python
fca_data = df.iloc[:,49:62]
```

```python
wcss = []
# 計算 k=1~10 的損失函數
for i in range(1,31):
    kmeans = cluster.KMeans(n_clusters=i, init='k-means++', random_state=0)
    kmeans.fit(fca_data)
    # kmeans.inertia_
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(9, 6)) 
plt.plot(range(1,31), wcss, marker='o')
plt.title('Elbow graph')
plt.xlabel('Cluster number')
plt.ylabel('WCSS')
plt.tight_layout()
plt.show()
```

```python
kmeans = cluster.KMeans(n_clusters=5, init='k-means++', random_state=0)
kmeans.fit(fca_data)

fca_data['clusters'] = kmeans.predict(fca_data)
centroids = round(pd.DataFrame(kmeans.cluster_centers_),2)
# fac_data

# random_state=0
```
